#Importando e Exportando dados com Python

##Importando o DF

###Preparando o ambiente

- Importando a biblioteca Pandas

```python
import pandas as pd
```

- Recebe o arquivo CarPrice_Assigment
- Cria um DataFrame sem cabeçalho

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(url, header = None)
```

- Apresenta as 5 primeiras linhas do dataframe

```python
df.head()
```

###Modificando informações no DF

- Define o novo nome das colunas

```python
headers = ["Symboling","Normalized-losses","Make","Fuel-type","Apiration","Num-of-doors","Body-style",
           "Drive-wheels","Engine-location","Wheel-base","Length","Width","Heigh","Curb-weight","Engine-type",
           "Num-of-cylinders","Engine-size","fuel-system","Bore","Stroke","Compression-ratio","Horsepower",
           "Peak-rpm","City-mpg","Highway-mpg","Price"]
df.columns = headers
```

- Apresenta as 5 primeiras linhas do novo DF

```python
df.head(5)
```

##Exportando o DF

- Salva o DF em um arquivo .csv

```python
path="/automobile.csv"
df.to_csv(path)
```
